## Import required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Data

## Hot Dog Contest Winners

```python
# Load data into dataframe
hdw_data = "ex1-2_datasets/hotdog-contest-winners.xlsm" 
hdw_df = pd.read_excel(hdw_data)

# Show data
hdw_df.head()
```

## Hot Dog Contest Placements

```python
# Load data into dataframe
hdp_data = "ex1-2_datasets/hotdog-places.xlsm" 
hdp_df = pd.read_excel(hdp_data)

# Show data
hdp_df.head()
```

## Obama Approval Ratings

```python
# Load data into dataframe
ob_data = "ex1-2_datasets/obama-approval-ratings.xls" 
ob_df = pd.read_excel(ob_data)

# Show data
ob_df.head()
```

## Bar Graph

```python
# Set figure and axes
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
fig.suptitle('Hot Dogs Eaten', fontsize=18, fontweight='bold')

# Horizontal Bar Graph
ax.barh(hdw_df.Year, hdw_df['Dogs eaten'])

# Set plot parameters
ax.set_xlabel('Number Hot Dogs Eaten', fontsize=14, fontweight='bold')
ax.set_ylabel('Year', fontsize=14, fontweight='bold')

# Show Plot
plt.show()
```

## Stacked Bar Graph

```python
import numpy as np

# Set figure and axes
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
fig.suptitle('Hot Dogs Eaten By Placement', fontsize=18, fontweight='bold')

# Position of the bars on the x-axis
N = len(hdp_df.columns)
r = np.arange(N)

# Heights of bars 1&2
bars = np.add(hdp_df.iloc[0], hdp_df.iloc[1]).tolist()

# Vertical Stacked Bar Graph
width = 0.4

p1 = plt.bar(r, hdp_df.iloc[0], width, color='b')
p2 = plt.bar(r, hdp_df.iloc[1], width, bottom=hdp_df.iloc[0], color='r')
p3 = plt.bar(r, hdp_df.iloc[2], width, bottom=bars, color='g')

# Set plot parameters
plt.xticks(r, list(hdp_df), fontweight='bold')
ax.set_xlabel('Year', fontsize=14, fontweight='bold')
ax.set_ylabel('Number Hot Dogs Eaten', fontsize=14, fontweight='bold')
header = ['1st', '2nd', '3rd']
plt.legend((p1[0], p2[0], p3[0]), (header[0], header[1], header[2]), fontsize=12, ncol=3, framealpha=0, fancybox=True)

# Show Plot
plt.show()
```

## Pie chart

```python
# Set figure and axes
fig1, ax1 = plt.subplots()
fig1.set_size_inches(6, 6)
fig1.suptitle('Obama Approval By Issue', fontsize=18, fontweight='bold')

# Set axes parameters
ax1.pie(ob_df['Approve'], 
        labels=ob_df.Issue, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show Plot
plt.show()
```

## Donut Chart

```python
# Set figure and axes
fig1, ax1 = plt.subplots()
fig1.set_size_inches(6, 6)
fig1.suptitle('Obama Approval By Issue', fontsize=18, fontweight='bold')

# Set axes parameters
ax1.pie(ob_df['Approve'], 
        labels=ob_df.Issue, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a circle at the center to turn pie into donut
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

# Show Plot
plt.show()
```
