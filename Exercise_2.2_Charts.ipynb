## Import required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load Data

```python
# Load data into dataframe
wpop_data = "ex2-2/world-population.xlsm" 
wpop_df = pd.read_excel(wpop_data)

# Show data
wpop_df.head()
```

## Line Chart

```python
# Set figure and axes
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
fig.suptitle('World Population', fontsize=18, fontweight='bold')

# Line Chart
ax.plot(wpop_df.Year, wpop_df.Population)

# Set plot parameters
ax.set_xlabel('Year', fontsize=14, fontweight='bold')
ax.set_ylabel('Population (Bilions)', fontsize=14, fontweight='bold')

# Set y-axis to start at 0
plt.ylim(0, 7000000000)

# Show Grid
ax.grid()

# Show Plot
plt.show()
```

## Step Chart

```python
# Set figure and axes
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
fig.suptitle('World Population', fontsize=18, fontweight='bold')

# Step Chart
# post steps up after value
plt.step(wpop_df.Year, wpop_df.Population, where='post')

# Set plot parameters
ax.set_xlabel('Year', fontsize=14, fontweight='bold')
ax.set_ylabel('Population (Bilions)', fontsize=14, fontweight='bold')

# Set y-axis to start at 0
plt.ylim(0, 7000000000)

# Show Grid
ax.grid()

# Show Plot
plt.show()
```
