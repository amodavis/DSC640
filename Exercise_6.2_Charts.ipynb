## Import required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
#import numpy as np
#from matplotlib import cm
```

# Load Data

```python
# Load Education data into dataframe
edu_data = "ex6-2/education.csv" 
edu_df = pd.read_csv(edu_data)

# Show data
edu_df.head()
```

```python
# Load SAT Scores into dataframe
sat_data = "ex6-2/tabn146.xls"

sat_df = pd.read_excel(sat_data, skiprows = 3)   # Start on the 4th row

# Drop records with null values
sat_df.dropna(axis=0, inplace=True)

# Drop 1st record w/ column numbers
sat_df.drop(sat_df.index[0], inplace=True)

# Rename columns
sat_df.columns = ["state",
    "reading_87_88",
    "math_87_88",
    "reading_95_96",
    "math_95_96",
    "reading_00_01",
    "math_00_01",
    "reading_05_06",
    "math_05_06",
    "writing_05_06",
    "reading_07_07",
    "math_07_08",
    "writing_07_08",
    "reading_08_09",
    "math_08_09",
    "writing_08_09",
    "pct_07_08",
    "pct_08_09"]

# Show data
sat_df.head(10)
```

## Histogram w/ matplotlib

```python
# Set plot area
fig, ax = plt.subplots()

# Build histogram
n, bins, patches = plt.hist(edu_df.reading, 20, density=1, facecolor='b', alpha=0.8)

# Set plot parameters
plt.xlabel('Reading Scores')
plt.ylabel('Probability')
plt.title('Histogram')

# Remove box frame.  Leave axes tics and labels only
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
left_side = ax.spines["left"]
left_side.set_visible(False)
bottom_side = ax.spines["bottom"]
bottom_side.set_visible(False)

# Save plot to file
plt.savefig('Py_histogram.png', bbox_inches='tight')

# Show plot
plt.show()
```

## Box Plot w/ matplotlib

```python
# Set plot area
fig, ax = plt.subplots()

# Build box plot
medianprops = dict(linestyle='-', linewidth=2.5, color='blue')
plt.boxplot(edu_df.reading, medianprops=medianprops)

# Set plot parameters
plt.title('Reading Scores')
plt.xticks([], [])

# Remove box frame.  Leave axes tics and labels only
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
left_side = ax.spines["left"]
left_side.set_visible(False)
bottom_side = ax.spines["bottom"]
bottom_side.set_visible(False)

# Save plot to file
plt.savefig('Py_boxplot.png', bbox_inches='tight')

# Show plot
plt.show()
```

## Bullet Chart w/ Seaborn

```python
import seaborn as sns

# Set plot area
fig, ax = plt.subplots()

# Pull records for plot
us_rec = sat_df.loc[sat_df['state'] == '   United States .....']
il_rec = sat_df.loc[sat_df['state'] == 'Illinois ......................']

# Set parameters for gauge bar
limits = [us_rec['reading_87_88'], 800]
palette = sns.color_palette("Blues_r", len(limits))

# Draw gauge bar 
prev_limit = 0
for idx, lim in enumerate(limits):
    ax.barh([1], lim-prev_limit, left=prev_limit, height=120, color=palette[idx])
    prev_limit = lim

# Select actual value to plot
data_to_plot = il_rec.reading_87_88
    
# Draw value bar
ax.barh([1], data_to_plot, color='black', height=30)

# Set plot parameters
ax.set_aspect('equal')
ax.set_yticks([1])
ax.set_yticklabels('Illinois')

plt.title('Reading Score Compared to US Average', loc='left', fontsize=18)
plt.ylabel('Illinois')

# Save plot to file
plt.savefig('Py_bullet.png', bbox_inches='tight')
```
